```python
def seasons(s):
    if (s==12 or s==1 or s==2):
        return 'summer'
    elif (s==3 or s==4 or s==5):
        return 'autumn'
    elif (s==6 or s==7 or s==8):
        return 'winter'
    else :
        return 'spring'
```

```python
weathdata['seasons']=weathdata['month'].apply(seasons)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

    
```

```python
weathdata['RainTomorrow'].value_counts()
```

```python
x,y=weathdata.drop(labels=['RainTomorrow'],axis=1),weathdata['RainTomorrow']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=23,test_size=0.2)
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
xtrain_sm,ytrain_sm=sm.fit_resample(xtrain,ytrain)
from collections import Counter
print('no of classes bf fit {}'.format(Counter(ytrain)))
print('no of classes af fit {}'.format(Counter(ytrain_sm)))
```

```python
model=RandomForestClassifier()
model.fit(xtrain_sm,ytrain_sm)
```

```python
predictions=model.predict(xtest)
```
